```python
import re

def parse_duration(duration):
    # ISO 8601 duration regex
    pattern = re.compile(
        r'PT(?:(?P<hours>\d+)H)?(?:(?P<minutes>\d+)M)?(?:(?P<seconds>\d+)S)?'
    )
    match = pattern.match(duration)
    if not match:
        return 0
    parts = match.groupdict()
    hours = int(parts['hours']) if parts['hours'] else 0
    minutes = int(parts['minutes']) if parts['minutes'] else 0
    seconds = int(parts['seconds']) if parts['seconds'] else 0
    return hours * 60 + minutes + seconds / 60
```

```python
df['Duration Minutes'] = df['Duration'].apply(parse_duration)
df['Duration Minutes'] 
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Duration Minutes'], bins=30, kde=True)
plt.title('Video Duration Distribution')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
def get_top_channels_per_country(df1, country, top_n=10):
    country_df = df1[df1['Country'] == country]
    top_channels = country_df['Channel Title'].value_counts().head(top_n)
    return top_channels
```

```python
countries = df1['Country'].unique()
countries
```

```python
fig, axes = plt.subplots(len(countries), 1, figsize=(10, len(countries) * 6))
for i, country in enumerate(countries):
    top_channels = get_top_channels_per_country(df1, country)
    sns.barplot(x=top_channels.values, y=top_channels.index, ax=axes[i], palette='viridis')
    axes[i].set_title(f'Top 10 Channels in {country}')
    axes[i].set_xlabel('Number of Videos')
    axes[i].set_ylabel('Channel Title')

plt.tight_layout()
plt.show()
```
